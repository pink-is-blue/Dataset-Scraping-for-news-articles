```python
import requests
from bs4 import BeautifulSoup
from htmldate import find_date
import datetime

# List of URLs to scrape
urls = ["https://www.politico.com/","https://indianexpress.com/section/political-pulse/"]
text_article=[]
link_article=[]
keyword = "politic"

for url in urls:
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all 'a' tags in the body of the HTML
        links = soup.find_all('a', href=True)
        
        # Extract and check if the link contains the keyword and associated text is greater than 35 characters
        for link in links:
            href = link['href']
            text = link.text.strip()
            
            if keyword in href.lower() and len(text) > 35:
                text_article.append(text)
                link_article.append(href)
                
                # Use htmldate to find the publication date of the URL
                pub_date_str = find_date(url)
                
                # Convert the publication date string to a datetime object
                try:
                    pub_date = datetime.datetime.strptime(pub_date_str, "%Y-%m-%d")
                    
                    # Check if the publication date is today
                    #if pub_date.date() == datetime.date.today():
                    print(f"Text: {text}\nLink: {href}\nPublication Date: {pub_date}\n")
                    
                except ValueError:
                    print(f"Failed to parse publication date for URL: {url}")
    else:
        print(f"Failed to retrieve the page for URL: {url}. Status code: {response.status_code}")
print(text_article)
print(link_article)
```

```python
from textblob import TextBlob

# List of sentences to analyze

from textblob import TextBlob

# Function to perform sentiment analysis on a list of sentences
def analyze_sentiment(text_article):
    sentiment_results = []
    for sentence in text_article:
        # Create a TextBlob object
        blob = TextBlob(sentence)
        # Get the sentiment polarity (-1 to 1: negative to positive)
        polarity = blob.sentiment.polarity
        # Classify the sentiment
        sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"
        # Append the results to the sentiment_results list in the format ["text", "sentiment"]
        sentiment_results.append([sentence, sentiment])
    return sentiment_results

# Example list of sentences to analyze

# Call the function and print the results
results = analyze_sentiment(text_article)
#for result in results:
    #print(f"Text: {result[0]}")
    #print(f"Sentiment: {result[1]}")
    #print("-" * 40)
print(results)

```
